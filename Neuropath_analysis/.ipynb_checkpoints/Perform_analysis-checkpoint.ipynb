# PRS & Pathology Spearman Correlation Processor

## Purpose
Computes pairwise Spearman rank correlation matrices between pathology features and Polygenic Risk Scores (PRS) for a cohort of donors. Utilizes a permutation-based resampling mechanism to derive robust empirical p-values and controls the False Discovery Rate (FDR) using Benjamini-Hochberg correction.

## Input
* `input/prs_table.csv`: Patient PRS data table.
* `input/neuropath_features.csv`: Parsed macromorphology pathology dataset.
* `diag_table`: Diagnostic reference data mapping donor labels.

> **Note:** The file paths listed above are indicative. Please replace them with your own local or relative file paths.

## Process
1. **Filter and align** donor cohorts across datasets to establish overlapping metrics.
2. **Compute** pairwise empirical Spearman rank correlations across observation masks.
3. **Resample** via random array shuffle permutations to create null distribution thresholds.
4. **Apply** multi-test FDR corrections and scale metric weights into plot dimensions.

## Output
* Saves an evaluation parameter payload data matrix to a binary pickle asset: `output/saved_plot_data_perm_10K(shade_imputed_apoe_correct).pkl`

```python
import os
import glob
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, zscore, tragedies, mannwhitneyu
from statsmodels.stats.multitest import multipletests

# Local workspace imports
from utils import feature_cols, shade_dict # predefined dicts
```

```python
# ==============================================================================
# MODULE 1: DATA ANALYSIS FUNCTIONS
# ==============================================================================

def prepare_correlation_data(
    path_df,
    prs_df,
    path_cols,
    prs_cols,
    donor_col="donor_id",
    prs_label_map=None,
    min_n=6,
    min_dot_size=50,
    max_dot_size=300,
    sig_threshold=0.05,
    star_scale=0.7,
    n_perm=1000,
    random_state=42,
):
    """
    Constructs statistical arrays, sizes, and matrices for visualization.
    """
    # Keep only columns that exist in the datasets
    path_cols = [c for c in path_cols if c in path_df.columns]
    prs_cols = [c for c in prs_cols if c in prs_df.columns]
    print(len(path_cols))

    # Inner join datasets on donor ID to ensure data alignment
    merged = pd.merge(
        path_df[[donor_col] + path_cols],
        prs_df[[donor_col] + prs_cols],
        on=donor_col,
        how="inner"
    ).drop_duplicates(donor_col)
    
    # Defensively force evaluation matrices to numeric format
    for col in path_cols + prs_cols:
        merged[col] = pd.to_numeric(merged[col], errors="coerce")

    corr_matrix = pd.DataFrame(index=path_cols, columns=prs_cols, dtype=float)
    pval_matrix = pd.DataFrame(index=path_cols, columns=prs_cols, dtype=float)

    rng = np.random.default_rng(random_state)

    # Core statistical processing loop using permuted null distributions
    for g_f in prs_cols:
        for p_f in path_cols:
            mask = merged[[p_f, g_f]].notna().all(axis=1).to_numpy()
            n_obs = int(mask.sum())

            # Insufficient observations filter
            if n_obs < min_n:
                corr_matrix.loc[p_f, g_f] = np.nan
                pval_matrix.loc[p_f, g_f] = 1.0
                continue

            x = merged.loc[mask, p_f].to_numpy()
            y = merged.loc[mask, g_f].to_numpy()

            # Variance boundary check
            if np.nanstd(x) == 0 or np.nanstd(y) == 0:
                corr_matrix.loc[p_f, g_f] = np.nan
                pval_matrix.loc[p_f, g_f] = 1.0
                continue

            obs_rho, _ = spearmanr(x, y)
            corr_matrix.loc[p_f, g_f] = obs_rho

            perm_rhos = np.empty(n_perm, dtype=float)

            # Generate empirical null distribution via permutation shuffling
            for i in range(n_perm):
                y_perm = rng.permutation(y)
                perm_rhos[i], _ = spearmanr(x, y_perm)

            p = (np.sum(np.abs(perm_rhos) >= abs(obs_rho)) + 1) / (n_perm + 1)
            pval_matrix.loc[p_f, g_f] = p

    # Execute global Benjamini-Hochberg FDR correction
    flat_p = pval_matrix.to_numpy().ravel()
    _, p_adj, _, _ = multipletests(flat_p, method="fdr_bh")

    padj_matrix = pd.DataFrame(
        p_adj.reshape(pval_matrix.shape),
        index=pval_matrix.index,
        columns=pval_matrix.columns
    )

    # Remap feature display names if dictionary mapping is provided
    if prs_label_map:
        new_cols = [prs_label_map.get(c, c) for c in prs_cols]
        corr_matrix.columns = new_cols
        pval_matrix.columns = new_cols
        padj_matrix.columns = new_cols
    else:
        new_cols = prs_cols

    # Calculate proportional marker size transformations
    logp = -np.log10(padj_matrix.replace(0, 1e-12)).fillna(0)
    lo = logp.min().min()
    hi = logp.max().max()

    if hi == lo:
        dot_size = pd.DataFrame(min_dot_size, index=logp.index, columns=logp.columns)
    else:
        scaled = (logp - lo) / (hi - lo)
        dot_size = min_dot_size + scaled * (max_dot_size - min_dot_size)

    # Map significant alpha layers and star rendering size markers
    sig_mask = (padj_matrix < sig_threshold).astype(int)
    star_size = sig_mask * dot_size * star_scale

    sample_counts = [merged[c].notna().sum() for c in path_cols]

    return {
        "corr_matrix": corr_matrix,
        "pval_matrix": pval_matrix,
        "padj_matrix": padj_matrix,
        "dot_size": dot_size,
        "sig_mask": sig_mask,
        "star_size": star_size,
        "sample_counts": sample_counts,
        "p_min": float(np.nanmin(p_adj)),
        "p_max": float(np.nanmax(p_adj)),
        "path_labels": path_cols,
        "prs_labels": new_cols,
    }


```

```python
# ==============================================================================
# MODULE 2: CONFIGURATION & CONFIGURABLE PARAMETERS
# ==============================================================================

# --- [INDICATIVE PATHS: Replace with your own project file paths] ---
INPUT_PRS_PATH = "output/input_prs_data.csv"
INPUT_MACRO_FEATURES_PATH = "input/input_pathology_features.csv"
OUTPUT_PICKLE_PATH = "output/output_results.pkl"

# Analysis threshold and execution parameters
MIN_OBSERVATIONS = 6
N_PERMUTATIONS = 10000
DONOR_ID_COLUMN = "donor_id"

# Metadata mapping and groupings, 
# It is recommended to create your own based on the names in the input file.
PRS_GROUP_NAME = 'Diseases'
PATHWAY_COLUMNS = feature_cols.copy() 
PRS_LABEL_MAP = shade_dict.copy() 


# ==============================================================================
# MODULE 3: PIPELINE EXECUTION & DATA CLEANING
# ==============================================================================

# Load cohort datasets using configured file paths
data_table = pd.read_csv(INPUT_PRS_PATH)
# UNCOMMENT TO AUDIT FILE LOADING:
# print(data_table.shape); data_table.head()

pred_df = pd.read_csv(INPUT_MACRO_FEATURES_PATH)
# UNCOMMENT TO AUDIT FILE LOADING:
# print(pred_df.shape); pred_df.head()

# Harmonize donor ids across cohorts
common_ids = compare_lists(data_table[DONOR_ID_COLUMN].to_list(), pred_df[DONOR_ID_COLUMN].to_list())

prs_df = data_table[data_table[DONOR_ID_COLUMN].isin(common_ids)]
path_df_temp = pred_df[pred_df[DONOR_ID_COLUMN].isin(common_ids)]
# UNCOMMENT TO MONITOR COHORT RETENTION COUNTS:
# print(prs_df.shape, path_df_temp.shape)

# Merge structural cohorts with clinical diagnosis variables
path_df = path_df_temp.merge(
    diag_table[[DONOR_ID_COLUMN, 'nd_current']],
    on=DONOR_ID_COLUMN,
    how="left"
)
# UNCOMMENT TO CHECK STRUCTURAL MERGE INTEGRITY:
# print(path_df.shape); path_df.head()

# Execute calculation routine using the configured parameters defined above
plot_data = prepare_correlation_data(
    path_df=path_df,
    prs_df=prs_df,
    path_cols=PATHWAY_COLUMNS,
    prs_cols=prs_cols,
    donor_col=DONOR_ID_COLUMN,
    prs_label_map=PRS_LABEL_MAP,
    min_n=MIN_OBSERVATIONS,
    n_perm=N_PERMUTATIONS
)
```
