# Permutation Data Plotter & Visualizer

## Purpose
Loads pre-computed permutation plot data from binary pickle assets, harmonizes data structures using local mapping dictionaries, and generates publication-ready correlation scattermaps (dot plots) for disease and shade clusters.

## Input
* Pre-computed pickle files containing correlation data configurations.
* Structural translation dictionaries mapped from local workspace scripts.

> **Note:** All file paths used in this process are indicative. Please replace them with your own local or relative file paths.

## Process
1. **Define** generic input/output file paths and visualization parameters at the top. 
*   [Figure: Dot Plots](#fig-dot-plots)
*   [Figure: Scatter Plots](#fig-scatter-plots)
*   [Figure: Heatmap - Feature Association](#fig-heatmap)
*   [Figure: Dot Plot - Pathology Load](#fig-pathology-load)

## Output
* Vector-based, publication-ready PDF figures saved to designated output locations.

```python
from utils.scattermap import scattermap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.stats import spearmanr, zscore

mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42

def plot_correlation_scattermap(
    plot_data: pd.DataFrame,          # The main DataFrame containing correlation metrics and coordinates
    bar_x: str = None,                # The column name in plot_data to use for the horizontal grouping/bar axis
    save_path: str = None,            # File path destination (including extension) to save the generated PDF/image
    h_bar: float = 0.07,              # Height of the horizontal structural bars relative to the total figure size
    v_bar: float = 0.01,              # Width of the vertical structural bars relative to the total figure size
    v_min: float = -0.5,               # Minimum correlation value bound for the color map scaling
    v_max: float = 0.5,                # Maximum correlation value bound for the color map scaling
    c_map: str = "Spectral_r",         # Color bar type
    cbar_title: str = "Correlation",
    fig_size: tuple = (25, 15)        # Total width and height dimensions of the output matplotlib figure in inches
):
    """
    Generates a scattermap plot with rank correlation, significance stars,
    and normalized blue bars on the side and top.
    """
    
    # --- Data Extraction & Initialization ---
    median_df_con_difference = plot_data["corr_matrix"]
    dot_size_df_log = plot_data["dot_size"]
    binary_stats_df = plot_data["sig_mask"]
    star_size = plot_data["star_size"]
    plot_diag_order = plot_data["path_labels"]
    prs_list = plot_data["prs_labels"]
    
    min_val = plot_data["p_min"]
    max_val = plot_data["p_max"]
    min_dot_size = 50
    max_dot_size = 300
    
#     plot_diag_order_clean = [rename_dict.get(name, name) for name in plot_data["path_labels"]]

#     # ── Rename matrices ──
#     median_df_con_difference = median_df_con_difference.rename(index=rename_dict)
#     dot_size_df_log          = dot_size_df_log.rename(index=rename_dict)
#     binary_stats_df          = binary_stats_df.rename(index=rename_dict)
#     star_size                = star_size.rename(index=rename_dict)

#     plot_diag_order = plot_diag_order_clean.copy()

    # --- Side Bar Plot Calculation (Y-axis) ---
    diagnosis_bar_df = pd.DataFrame({
        "Diagnosis": plot_diag_order,
        "Number_of_Observations": plot_data["sample_counts"]
    })
    raw_counts = diagnosis_bar_df["Number_of_Observations"].values
    #left_span = max(2, len(prs_list) * 0.25)
    left_span = max(2, len(prs_list) * h_bar)#current value was 0.25
    diagnosis_bar = -(raw_counts / max(raw_counts.max(), 1)) * left_span
    diag_pos = np.arange(len(plot_diag_order)) + 0.5

    # --- Top Bar Plot Calculation (X-axis) ---
    ##top_bar_counts = np.array([bar_x.get(prs, 0) for prs in prs_list])##edited for plotting new data
    
    if bar_x is None:
        bar_x = plot_data.get("bar_x", {})

    top_bar_counts = np.array([bar_x.get(prs, 0) for prs in prs_list])
    
    # REDUCED SCALE: Changed multiplier from 0.25 to 0.1 for a smaller "big" bar
    #top_span = max(1, len(plot_diag_order) * 0.1) 
    top_span = max(1, len(plot_diag_order) * v_bar) #current value was 0.1
    
    top_bar_heights = -(top_bar_counts / max(top_bar_counts.max(), 1)) * top_span
    prs_pos = np.arange(len(prs_list)) + 0.5

    # --- Plot Styling ---
    sns.set(style="darkgrid", font_scale=2, 
            rc={'axes.facecolor':'#F0E6EB', "grid.linestyle": "-", "grid.color": '#b0b0b0'})
    
    fig = plt.figure(figsize=fig_size, dpi=200)

    ax1 = plt.subplot2grid((9, 8), (0, 0), colspan=4, rowspan=9) 
    ax2 = plt.subplot2grid((9, 8), (0, 4), colspan=1, rowspan=1) 
    ax3 = plt.subplot2grid((9, 8), (3, 4), colspan=1, rowspan=1) 

    # --- Main Scattermap ---
    ax1 = scattermap(median_df_con_difference,
                     cmap=sns.color_palette(c_map, as_cmap=True),
                     marker_size=dot_size_df_log,
                     ax=ax1,
                     vmin=v_min,#-0.5,
                     vmax=v_max,#0.5,
                     linecolor='black',
                     linewidths=0.2,
                     cbar=True,
                     cbar_kws={"shrink": .5})

    # --- ADDED THESE TWO LINES HERE TO FIX THE LABEL ---
    cbar = ax1.collections[0].colorbar
    cbar.ax.text(-0.2, 1.05, cbar_title, 
                 ha='left', va='bottom', fontsize=14, transform=cbar.ax.transAxes)
    cbar.ax.tick_params(labelsize=14)
    # --- Overlay Significance Stars ---
    star_cmap = mcolors.ListedColormap(["#E59400"])
    ax1 = scattermap(binary_stats_df,
                     marker='*',
                     marker_size=star_size,
                     cbar=False,
                     ax=ax1,
                     linecolor='black',
                     linewidths=0.2,
                     cmap=star_cmap)

    # --- Bars (Color matched to steelblue) ---
    bar_color = 'steelblue'
    ax1.barh(diag_pos, diagnosis_bar, 0.6, alpha=1, color=bar_color, edgecolor="none")
    ax1.bar(prs_pos, top_bar_heights, 0.6, alpha=1, color=bar_color, edgecolor="none")

    # --- Axis Formatting & Limits ---
    ax1.xaxis.tick_top() 
    ax1.tick_params('x', labelrotation=90)
    ax1.tick_params('y', labelrotation=0)
    
    # Adjust limits for the new smaller span
    ax1.set_xlim([min(diagnosis_bar) - 0.5, len(prs_list)])
    ax1.set_ylim([len(plot_diag_order), min(top_bar_heights) - 0.5]) 
    
    # Zero lines
    ax1.axvline(x=0, color='k', linewidth=1)
    ax1.axhline(y=0, color='k', linewidth=1)
    ax1.set_ylabel('')
    ax1.set_xlabel('')

    # --- Legend Tricks ---
    sc = ax2.scatter([2, 5], [2, 5], s=[min_dot_size, max_dot_size], alpha=0.5, c='white')
    L = ax2.legend(*sc.legend_elements("sizes"), loc='center right', bbox_to_anchor=(1, 0.7), frameon=False)
    L.set_title("FDR corrected p-value", prop={'size': 12})
    L.get_texts()[1].set_text(round(min_val, 2))
    L.get_texts()[0].set_text(round(max_val, 2))
    ax2.axis('off')

    ax3.scatter([4], [2], s=[max_dot_size], alpha=1, c='#E59400', marker='*')
    ax3.legend(['significance'], loc='center right', bbox_to_anchor=(1, 1), frameon=False)
    ax3.axis('off')

    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
    
    plt.show()
    

    


def plot_prs_vs_pathology(
    prs_df, path_df, prs_col, path_col, 
    save_path=None, rename_dict=None,
    xticks=None, yticks=None, xlabel=None, ylabel=None,
    figsize=(4, 5), dpi=300, 
    use_zscore=False,
    **kwargs
):
    # --- PDF Editability ---
    #mpl.rcParams['pdf.fonttype'] = 42
    #mpl.rcParams['ps.fonttype'] = 42

    # 1. Merge & Clean
    merged_df = pd.merge(prs_df[['donor_id', prs_col]], path_df[['donor_id', path_col]], on='donor_id').dropna()
    merged_df[path_col] = pd.to_numeric(merged_df[path_col], errors='coerce')
    merged_df = merged_df.dropna()
    if use_zscore:
        merged_df[prs_col] = zscore(merged_df[prs_col], nan_policy="omit")

    # 2. Stats
    rho, p_val = spearmanr(merged_df[path_col], merged_df[prs_col])
    neg_log_p = -np.log10(p_val) if p_val > 0 else 0
    stats_text = f"ρ = {rho:.2f}\n[-log10P = {neg_log_p:.2f}]"

    # 3. Setup Plot
    sns.set_style("white")
    fig, ax = plt.subplots(figsize=figsize, dpi=dpi)
    
    # 4. Violin plot with trendline (Updated with original blue color defaults)
    # Draw violins first
    sns.violinplot(
        data=merged_df, x=path_col, y=prs_col, ax=ax,
        color=kwargs.get('pt_color', "#9ed2ec"), alpha=0.4, # Restored to light blue
        inner="box", linewidth=1.5
    )
    
    # Overlay the regression trendline (convert x-axis categories to internal numeric indices [0, 1, 2...])
    unique_x = sorted(merged_df[path_col].unique())
    x_mapping = {val: idx for idx, val in enumerate(unique_x)}
    merged_df['_x_idx'] = merged_df[path_col].map(x_mapping)
    
    sns.regplot(
        data=merged_df, x='_x_idx', y=prs_col, ax=ax,
        scatter=False, # Suppress scatter points entirely
        line_kws={
            'color': kwargs.get('line_color', "#74c2e6"), # Restored to darker blue trendline
            'lw': 4
        }
    )

    # 5. Safe Text Placement
    ax.text(
        0.05, 0.95, stats_text, 
        transform=ax.transAxes, 
        fontsize=14, weight='bold', color='black',
        verticalalignment='top',
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=3.0)
    )

    # 6. Labels & Ticks
    path_label = rename_dict.get(path_col, path_col) if rename_dict else path_col
    prs_label = rename_dict.get(prs_col, prs_col) if rename_dict else prs_col

    ax.set_xlabel(xlabel or path_label, fontsize=18, weight='bold')
    ax.set_ylabel(ylabel or f"{prs_label} (PRS)", fontsize=18, weight='bold')
    
    # Fix Category tick strings/values to match original positions
    ax.set_xticks(range(len(unique_x)))
    ax.set_xticklabels(xticks if xticks is not None else unique_x)
    
    if yticks is not None:
        ax.set_yticks(yticks)
    else:
        ax.yaxis.set_major_locator(plt.MaxNLocator(nbins=5))
    
    ax.tick_params(axis='both', which='both', reset=True, direction='out', length=6, width=1.5, labelsize=16, top=False, right=False)
    sns.despine(ax=ax, top=True, right=True, left=False, bottom=False)
    plt.tight_layout()

    if save_path:
        fig.savefig(save_path, format="pdf", bbox_inches='tight')
    plt.show()
```

## <a id="fig-pathology-load"></a>Figure: Dot Plot - Pathology Load

```python
from scipy.stats import fisher_exact, mannwhitneyu
from statsmodels.stats.multitest import multipletests
import numpy as np
import pandas as pd


def _calculate_feature_burden(df, feature, diagnosis_keep, control_label, binary_cols, min_n, group_sizes):
    """Calculates n, group means, donor proportions, differences, and raw p-values for a single feature as fractions."""
    results = {"n": {}, "prop": {}, "mean": {}, "diff": {}, "pval": {}}
    
    control_vals = df.loc[df["_dx"] == control_label, feature].dropna()
    is_binary = feature in binary_cols
    # Removed * 100 to keep as a fraction
    control_summary = (control_vals == 1).mean() if is_binary else control_vals.mean() if len(control_vals) else np.nan

    for dx in diagnosis_keep:
        vals = df.loc[df["_dx"] == dx, feature].dropna()
        n_len = len(vals)
        denom = group_sizes.loc[dx]
        
        results["n"][dx] = n_len
        # Removed * 100 to keep as a fraction
        results["prop"][dx] = (n_len / denom) if denom > 0 else np.nan
        
        # Removed * 100 to keep as a fraction
        group_summary = (vals == 1).mean() if is_binary else vals.mean() if n_len else np.nan
        results["mean"][dx] = group_summary
        results["diff"][dx] = 0.0 if dx == control_label else (group_summary - control_summary)

        if dx == control_label:
            results["pval"][dx] = np.nan
        elif n_len < min_n or len(control_vals) < min_n:
            results["pval"][dx] = 1.0
        elif is_binary:
            a, b = int((vals == 1).sum()), int((vals == 0).sum())
            c, d = int((control_vals == 1).sum()), int((control_vals == 0).sum())
            _, p = fisher_exact([[a, b], [c, d]])
            results["pval"][dx] = p
        else:
            combined = pd.concat([vals, control_vals])
            if combined.nunique() < 2:
                results["pval"][dx] = 1.0
            else:
                _, p = mannwhitneyu(vals, control_vals, alternative="two-sided")
                results["pval"][dx] = p
                
    return results


def prepare_pathology_burden_by_diagnosis(
    path_df,
    path_cols,
    diagnosis_col="nd_current",
    diagnosis_keep=("Control", "AD", "PD"),
    control_label="Control",
    rename_dict=None,
    binary_cols=None,
    scale_max=None,
    min_n=6,
    min_dot_size=80,
    max_dot_size=320,
    sig_threshold=0.05,
    star_scale=0.7,
    normalize="known_scale",
):
    df = path_df.copy()

    amyloid_map = {"O": 0, "A": 1, "B": 2, "C": 3}
    for col in ["Braak_score_amyloid", "A", "B", "C"]:
        if col in df.columns:
            df[col] = df[col].replace(amyloid_map)

    path_cols = [c for c in path_cols if c in df.columns]
    binary_cols = set(binary_cols or [])
    scale_max = scale_max or {}

    diagnosis_keep = [d for d in diagnosis_keep if d in df[diagnosis_col].dropna().unique()]
    if control_label not in diagnosis_keep:
        raise ValueError(f"{control_label} not found in diagnosis groups.")

    df = df[df[diagnosis_col].isin(diagnosis_keep)].copy()
    df["_dx"] = df[diagnosis_col]

    for col in path_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    group_sizes = df["_dx"].value_counts().reindex(diagnosis_keep).fillna(0)

    mean_matrix = pd.DataFrame(index=path_cols, columns=diagnosis_keep, dtype=float)
    diff_matrix = pd.DataFrame(index=path_cols, columns=diagnosis_keep, dtype=float)
    pval_matrix = pd.DataFrame(index=path_cols, columns=diagnosis_keep, dtype=float)
    n_matrix = pd.DataFrame(index=path_cols, columns=diagnosis_keep, dtype=float)
    prop_matrix = pd.DataFrame(index=path_cols, columns=diagnosis_keep, dtype=float)

    # 1. Pathology Burden Extraction Loop
    for feature in path_cols:
        res = _calculate_feature_burden(df, feature, diagnosis_keep, control_label, binary_cols, min_n, group_sizes)
        mean_matrix.loc[feature] = res["mean"]
        diff_matrix.loc[feature] = res["diff"]
        pval_matrix.loc[feature] = res["pval"]
        n_matrix.loc[feature] = res["n"]
        prop_matrix.loc[feature] = res["prop"]

    # 2. Result Compilation and Normalization
    plot_matrix = diff_matrix.copy()
    if normalize == "known_scale":
        for feature in path_cols:
            if feature not in binary_cols:
                max_possible = scale_max.get(feature, None)
                if max_possible is not None and max_possible > 0:
                    plot_matrix.loc[feature] = diff_matrix.loc[feature] / max_possible

    # FDR correction
    non_control_cols = [d for d in diagnosis_keep if d != control_label]
    test_pvals = pval_matrix[non_control_cols].to_numpy().ravel()

    if len(test_pvals) > 0:
        _, p_adj, _, _ = multipletests(test_pvals, method="fdr_bh")
        padj_matrix = pd.DataFrame(
            p_adj.reshape(pval_matrix[non_control_cols].shape),
            index=pval_matrix.index,
            columns=non_control_cols,
        )
        padj_matrix[control_label] = 1.0
        padj_matrix = padj_matrix[diagnosis_keep]
    else:
        padj_matrix = pval_matrix.fillna(1.0)

    pval_matrix[control_label] = 1.0

    # Dot scaling calculations (Now based on fraction scale 0.0 - 1.0)
    size_values = prop_matrix.copy()
    lo, hi = np.nanmin(size_values.values), np.nanmax(size_values.values)

    if hi == lo:
        dot_size = pd.DataFrame(min_dot_size, index=size_values.index, columns=size_values.columns)
    else:
        dot_size = min_dot_size + ((size_values - lo) / (hi - lo)) * (max_dot_size - min_dot_size)

    sig_mask = (padj_matrix < sig_threshold).astype(int)
    sig_mask[control_label] = 0
    star_size = sig_mask * dot_size * star_scale

    if rename_dict:
        plot_matrix = plot_matrix.rename(index=rename_dict)
        mean_matrix = mean_matrix.rename(index=rename_dict)
        diff_matrix = diff_matrix.rename(index=rename_dict)
        pval_matrix = pval_matrix.rename(index=rename_dict)
        padj_matrix = padj_matrix.rename(index=rename_dict)
        dot_size = dot_size.rename(index=rename_dict)
        sig_mask = sig_mask.rename(index=rename_dict)
        star_size = star_size.rename(index=rename_dict)
        n_matrix = n_matrix.rename(index=rename_dict)
        prop_matrix = prop_matrix.rename(index=rename_dict)

    path_labels = plot_matrix.index.tolist()
    diagnosis_labels = plot_matrix.columns.tolist()
    bar_x = df["_dx"].value_counts().reindex(diagnosis_labels).fillna(0).to_dict()
    sample_counts = [df[c].notna().sum() for c in path_cols]

    return {
        "corr_matrix": plot_matrix,
        "mean_raw_matrix": mean_matrix,
        "diff_matrix": diff_matrix,
        "raw_pval_matrix": pval_matrix,
        "padj_matrix": padj_matrix,
        "dot_size": dot_size,
        "sig_mask": sig_mask,
        "star_size": star_size,
        "sample_counts": sample_counts,
        "p_min": float(np.nanmin(prop_matrix.values)),
        "p_max": float(np.nanmax(prop_matrix.values)),
        "n_matrix": n_matrix,
        "prop_matrix": prop_matrix,
        "dot_size_legend_title": "Donor proportion (%)",
        "path_labels": path_labels,
        "prs_labels": diagnosis_labels,
        "bar_x": bar_x,
    }
```
